```python
import cv2
import matplotlib.pyplot as plt
from src.data.transforms import seeTransforms
```

Define pipeline

```python
traintxt = "/content/drive/MyDrive/Lane_Detection_Project/data/tusimple/train.txt"

with open(traintxt, "r") as f:
    lines = f.readlines()


ROOT = "/content/drive/MyDrive/Lane_Detection_Project/data/tusimple/"
image_path, bin_path, inst_path = lines[0].strip().split()

image = cv2.imread(f"{ROOT}{image_path}")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
binary_mask = cv2.imread(f"{ROOT}{bin_path},", cv2.IMREAD_GRAYSCALE)
instance_mask = cv2.imread(f"{ROOT}{inst_path}", cv2.IMREAD_GRAYSCALE)

augmented = seeTransforms(f"{ROOT}{image_path}", f"{ROOT}{bin_path}", f"{ROOT}{inst_path}")

aug_image = augmented[0]
aug_binary = augmented[1]
aug_instance = augmented[2]


# Show results
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

# Original row
axes[0,0].imshow(image)
axes[0,0].set_title("Original Image")

axes[0,1].imshow(binary_mask, cmap="gray")
axes[0,1].set_title("Original Binary Mask")

axes[0,2].imshow(instance_mask)
axes[0,2].set_title("Original Instance Mask")

# Augmented row
axes[1,0].imshow(aug_image)
axes[1,0].set_title("Augmented Image")

axes[1,1].imshow(aug_binary, cmap="gray")
axes[1,1].set_title("Augmented Binary Mask")

axes[1,2].imshow(aug_instance)
axes[1,2].set_title("Augmented Instance Mask")

for ax in axes.flatten():
    ax.axis("off")

plt.tight_layout()
plt.show()



```
